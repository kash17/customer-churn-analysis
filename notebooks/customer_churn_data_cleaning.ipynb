```python
#check missing values

df.isnull().sum()
```

```python
#check duplicates

df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.shape
```

```python
#strip spaces for column values

for col in df.select_dtypes(include='object').columns:
    df[col]=df[col].str.strip()
```

```python
df.dtypes
```

```python
#convert customer id to string

df['CustomerID']=df['CustomerID'].astype('string')
```

```python
#ensure numeric columns

# Convert numeric columns safely
df['TenureMonths'] = pd.to_numeric(df['TenureMonths'], errors='coerce')
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')

# Handle missing values
df['TenureMonths'] = df['TenureMonths'].fillna(df['TenureMonths'].median()).astype('int64')
df['MonthlyCharges'] = df['MonthlyCharges'].fillna(df['MonthlyCharges'].median())
```

```python
#convert categorical columns

categorical_col= ['ContractType','InternetService','PaymentMethod','Churn']

for col in categorical_col:
    df[col]=df[col].astype('category')
    df[col].fillna(df[col].mode(),inplace=True)
```

```python
# Create Churn_Flag safely from messy data
df['Churn_Flag'] = (
    df['Churn']
    .astype(str)        # handle NaN and non-string values
    .str.strip()        # remove extra spaces
    .str.lower()        # normalize casing
    .eq('yes')          # True if churn == 'yes'
    .astype('int8')     # convert True/False to 1/0
)
```

```python
# Check if any values failed to map
df[df['Churn_Flag'].isna()]['Churn'].unique()
```
