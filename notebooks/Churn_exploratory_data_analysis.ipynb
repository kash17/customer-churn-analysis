```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:\Karishma\Data analysis classes\projects\customer churn analysis\cleaned_Customer_Churn_Dataset.csv")
```

```python
df.head()
```

```python
#overall churn rate

churn_rate=df['Churn_Flag'].value_counts(normalize=True)*100
churn_rate.plot(kind='bar')
plt.title("overall customer churn rate (%)")
plt.xlabel("churn flag (0=no 1=yes)")
plt.ylabel("Percentage")
plt.show()
```

```python
#Churn by Contract Type

contract_churn=df.groupby('ContractType')['Churn_Flag'].mean()
contract_churn.plot(kind='bar')
plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

```python
#Churn by Internet Service

internet_churn = df.groupby('InternetService')['Churn_Flag'].mean()
internet_churn.plot(kind='bar')
plt.title("Churn Rate by Internet Service")
plt.ylabel("Churn Rate")
plt.show()
```

```python
#Tenure vs Churn Trend

tenure_churn = df.groupby('TenureMonths')['Churn_Flag'].mean()
plt.figure(figsize=(10,4))
tenure_churn.plot()
plt.title("Churn Trend by Customer Tenure")
plt.xlabel("Tenure (Months)")
plt.ylabel("Churn Rate")
plt.show()
```

```python
#Monthly Charges vs Churn

df.boxplot(column='MonthlyCharges', by='Churn')
plt.title("Monthly Charges vs Churn")
plt.suptitle("")
plt.ylabel("Monthly Charges")
plt.show()
```

```python
#Payment Method vs Churn

payment_churn=df.groupby('PaymentMethod')['Churn_Flag'].mean()
payment_churn.plot(kind='bar')
plt.title("Churn Rate by Payment Method")
plt.ylabel("Churn rate")
plt.show()
```

```python
#Revenue Loss Due to Churn

monthly_revenue_loss=df[df['Churn_Flag']==1]['MonthlyCharges'].sum()

monthly_revenue_loss #Churned customers contribute 562513.77 ₹ monthly revenue loss.
```

```python
#Final Business Insights

# Key Insights
#Month-to-month contract customers have the highest churn
- Customers with tenure < 12 months are high-risk
- High monthly charges significantly increase churn
- Certain payment methods show higher churn patterns

# Business Recommendations
- Promote long-term contracts with discounts
- Improve first 3-month onboarding experience
- Offer retention offers to high-value churn-risk customers
```
